Load data into dictionary

```python
import json
path = '../output/data.json'
d = json.load(open(path))['result']
```

Create `pandas` dataframes

```python
import pandas as pd
counts, points = d['counts'], d['points']
green_rows = [[t, 'green', i, j] for t, (i, j) in enumerate(zip(counts['2'], points['2']))]
red_rows = [[t, 'red', i, j] for t, (i, j) in enumerate(zip(counts['4'], points['4']))]
rows = green_rows+red_rows
df = pd.DataFrame(rows, columns=['time', 'type', 'counts', 'points'])
```

Vizualize results

```python
import seaborn as sns

# counts of cells over time
sns.lineplot(data=df, x='time', y='counts', hue='type')
```

```python
sns.lineplot(data=df, x='time', y='points', hue='type')
```
