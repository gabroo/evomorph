Load data into dictionary and create `pandas` dataframe

```python
import json
import pandas as pd

path = '../output/data.json'
d = json.load(open(path))['result']
df = pd.DataFrame(d, columns=['time', 'type', 'points'])
```

Vizualize results

```python
import seaborn as sns

averages = pd.DataFrame(df.groupby(['time', 'type']).mean().to_records())
std = pd.DataFrame(df.groupby(['time', 'type']).std().to_records())
```

```python
ax = sns.lineplot(data=averages, x='time', y='points', hue='type')
ax.axhline(5263, ls='--', c='gray')
```

```python
ax = sns.lineplot(data=std, x='time', y='points', hue='type')
```

```python
counts = pd.DataFrame(df.groupby(['time', 'type']).count().to_records())
```

ax = sns.lineplot(data=counts, x='time', y='points', hue='type')
